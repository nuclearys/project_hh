3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**.

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…
    * Опыт работы 11 месяцев…
    * Учитывайте эту особенность в вашем коде

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
import pandas as pd
import numpy as np
import re

def extract_work_experience(experience_string):
    if pd.isna(experience_string) or experience_string == 'Не указано':
        return np.nan

    # Удаляем лишнюю информацию, оставляем только опыт
    match = re.match(r'Опыт работы (.*?)(?:,|$)', experience_string)

    if match:
        experience = match.group(1)

        years = re.search(r'(\d+)\s*(?:год|года|лет)', experience)
        months = re.search(r'(\d+)\s*(?:месяц|месяца|месяцев)', experience)

        total_months = 0

        # Проверяем присутствие лет и месяцев
        if years:
            total_months += int(years.group(1)) * 12
        if months:
            total_months += int(months.group(1))

        return int(total_months) if total_months > 0 else np.nan

    return np.nan  # Возврат NaN, если нет соответствия

# Применение функции и создание нового столбца
student_data['Опыт работы (месяц)'] = student_data['Опыт работы'].apply(extract_work_experience)
```

Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
million_cities = [
    'Новосибирск', 'Екатеринбург', 'Нижний Новгород', 'Казань',
    'Челябинск', 'Омск', 'Самара', 'Ростов-на-Дону',
    'Уфа', 'Красноярск', 'Пермь', 'Воронеж', 'Волгоград'
]

def categorize_city(city):
    """Определяет категорию города."""
    if city == 'Москва':
        return 'Москва'
    elif city == 'Санкт-Петербург':
        return 'Санкт-Петербург'
    elif city in million_cities:
        return 'город-миллионник'
    else:
        return 'другие'

def check_relocate_readiness(text):
    text = text.lower()
    phrase = 'не'
    # Проверяем, содержится ли фраза в предложении
    return not(phrase in text)


def check_relocate_trip(text):
    text = text.lower()
    phrase = 'не'
    # Проверяем, содержится ли фраза в предложении
    return not(phrase in text)

def extract_city_relocate_trip(city_relocate_trip_string):
    #parts = city_relocate_trip_string.split(', ')
    parts = re.split(r'\s*,\s*', city_relocate_trip_string)
    city = parts[0].strip()
    relocate = parts[1].strip()
    trip = parts[2].strip()
    # Проверяем всевозмодные вварианты для трех значений

    if parts[0].strip() == '':
      city == None
    elif parts[0].strip() != '':
      city == parts[0].strip()
    elif parts[1].strip() == '':
      relocate == None
    elif parts[1].strip() != '':
      relocate == parts[1].strip()
    elif parts[2].strip() == '':
      trip == None
    elif parts[2].strip() != '':
      trip == parts[2].strip()

    # Категоризация города
    categorized_city = categorize_city(city)
    relocate_readiness = check_relocate_readiness(relocate)
    relocate_trip = check_relocate_trip(trip)

    return pd.Series([categorized_city, relocate_readiness, relocate_trip])

# Применение функции и создание новых столбцов
student_data[['Город', 'Готовность к переезду', 'Готовность к командировкам']] = student_data['Город, переезд, командировки'].apply(extract_city_relocate_trip)
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
# Разделение признаков ддля столбца "Занятость"

import re

def extract_business(business):
    parts = re.split(r'\s*,\s*', business)
    full_time = False
    part_time = False
    project_work = False
    stager_work = False
    volontree_work = False

    for part in parts:
        if part == 'полная занятость':
            full_time = True
        elif part == 'частичная занятость':
            part_time = True
        elif part == 'проектная работа':
            project_work = True
        elif part == 'стажировка':
            stager_work = True
        elif part == 'волонтерство':
            volontree_work = True

    return pd.Series([full_time, part_time, project_work, stager_work, volontree_work])

# Применение функции и создание новых столбцов
student_data[['полная занятость', 'частичная занятость', 'проектная работа', 'стажировка', 'волонтерство']] = student_data['Занятость'].apply(extract_business)
```

```python
# Разделение признаков ддля столбца "График"

import re

def extract_shedule(shedule):
    parts = re.split(r'\s*,\s*', shedule)
    remote_work = False
    flexible_work = False
    full_day = False
    shift_schedule = False
    shift_method = False

    for part in parts:
        if part == 'удаленная работа':
          remote_work = True
        elif part == 'гибкий график':
          flexible_work = True
        elif part == 'полный день':
          full_day = True
        elif part == 'сменный график':
          shift_schedule = True
        elif part == 'вахтовый метод':
          shift_method = True

    return pd.Series([remote_work, flexible_work, full_day, shift_schedule, shift_method])

# Применение функции и создание новых столбцов
student_data[['удаленная работа', 'гибкий график', 'полный день', 'сменный график', 'вахтовый метод']] = student_data['График'].apply(extract_shedule)
```

Теперь у нас есть вся необходимая информация для того, чтобы создать признак "ЗП (руб)" - заработная плата в рублях.

После ответа на контрольные вопросы удалите исходный столбец заработной платы "ЗП" и все промежуточные столбцы, если вы их создавали.

Итак, давайте обсудим возможный алгоритм преобразования:
1. Перевести признак "Обновление резюме" из таблицы с резюме в формат datetime и достать из него дату. В тот же формат привести признак "date" из таблицы с валютами.
2. Выделить из столбца "ЗП" сумму желаемой заработной платы и наименование валюты, в которой она исчисляется. Наименование валюты перевести в стандарт ISO согласно с таблицей выше.
3. Присоединить к таблице с резюме таблицу с курсами по столбцам с датой и названием валюты (подумайте, какой тип объединения надо выбрать, чтобы в таблице с резюме сохранились данные о заработной плате, изначально представленной в рублях). Значение close для рубля заполнить единицей 1 (курс рубля самого к себе)
4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
#Импортирование данных о курсах валют
import pandas as pd
exchange_data = pd.read_csv('/content/drive/MyDrive/ExchangeRates.csv', sep=',', on_bad_lines='skip')
```

```python
import pandas as pd
student_data['Обновление резюме'] = pd.to_datetime(student_data['Обновление резюме'], errors='coerce')
```

```python
student_data['Обновление резюме'] = student_data['Обновление резюме'].dt.normalize().dt.strftime('%Y-%m-%d')
```

```python
import re

def extract_salary(salary):
    salary_vol, currency = salary.split()
    proportion = 0

    if currency == 'грн.':
        currency = 'UAH'
        proportion = 10
    elif currency == 'USD':
        currency = 'USD'
        proportion = 1
    elif currency == 'EUR':
        currency = 'EUR'
        proportion = 1
    elif currency == 'белруб':
        currency = 'BYN'
        proportion = 1
    elif currency == 'KGS':
        currency = 'KGS'
        proportion = 10
    elif currency == 'сум':
        currency = 'UZS'
        proportion = 10000
    elif currency == 'AZN':
        currency = 'AZN'
        proportion = 1
    elif currency == 'KZT':
        currency = 'KZT'
        proportion = 100
    elif currency == 'руб.':
        currency = 'руб.'
        proportion = 1

    return pd.Series([salary_vol, currency, proportion])

# Применение функции и создание новых столбцов
student_data[['Зарплата', 'Валюта в ISO', 'Пропорция']] = student_data['ЗП'].apply(extract_salary)
```

```python
# Переименование столбца в student_data
exchange_data = exchange_data.rename(columns={'date': 'Обновление резюме'})
```

```python
exchange_data = exchange_data.rename(columns={'currency': 'Валюта в ISO'})
```

```python
df_combined = pd.merge(student_data, exchange_data, on=['Обновление резюме', 'Валюта в ISO'])
```

```python
combined_df = pd.concat([student_data[~student_data['ЗП'].str.contains('AZN|KZT|USD|EUR|грн.|сум|KGS')], df_combined], ignore_index=True)
```

```python
combined_df['Зарплата'] =combined_df['Зарплата'].astype(float)
combined_df['proportion'] = combined_df['proportion'].astype(float)
```

```python
combined_df['ЗП(руб)'] = ((combined_df['close'] * combined_df['Зарплата'])/(combined_df['proportion']))
```

```python
# Анализ распределения по возрасту соискателей
import matplotlib.pyplot as plt
import seaborn as sns

# Смотрим основные статистики
age_stats = combined_df['Возраст'].describe()

# Мода
age_mode =combined_df['Возраст'].mode()[0]

# Предельные значения
age_min = age_stats['min']
age_max = age_stats['max']

# Вычисляем интервал для большинства соискателей
age_intervals = (age_stats['25%'], age_stats['75%'])

# Графическое представление распределения
plt.figure(figsize=(10, 6))
sns.histplot(combined_df['Возраст'], bins=range(int(age_min), int(age_max) + 1), kde=True)
plt.title('Распределение по возрасту соискателей')
plt.xlabel('Возраст')
plt.ylabel('Частота')
plt.show()
```

```python
# Анализ аномалий
# Определяем аномалии
age_mean = age_stats['mean']
age_std = age_stats['std']
lower_bound = age_mean - 2 * age_std
upper_bound = age_mean + 2 * age_std

anomalies = combined_df[(combined_df['Возраст'] < lower_bound) | (combined_df['Возраст'] > upper_bound)]

# Печатаем результаты
print(f"Мода возраста: {age_mode}")
print(f"Минимальный возраст: {age_min}")
print(f"Максимальный возраст: {age_max}")
print(f"Интервал возраста большинства соискателей: {age_intervals}")
print(f"Количество аномалий в возрасте: {len(anomalies)}")
print(f"Аномальные значения возраста:\n{anomalies['Возраст'].unique()}")

```

```python
combined_df = combined_df[combined_df['Возраст'] > 0]

# Логарифмирование возраста
combined_df['log_age'] = np.log(combined_df['Возраст'])

# Среднее и стандартное отклонение
mean_log_age = combined_df['log_age'].mean()
std_log_age = combined_df['log_age'].std()

# Устанавливаем границы интервала методом трех сигм
lower_bound = mean_log_age - 3 * std_log_age
upper_bound = mean_log_age + 3 * std_log_age

# Визуализация
plt.figure(figsize=(10, 6))
histplot = sns.histplot(combined_df['log_age'], bins=10, kde=True)
histplot.set_title('Распределение возраста в логарифмическом масштабе')
histplot.set_xlabel('Логарифм возраста')

# Добавление линий
histplot.axvline(mean_log_age, color='k', lw=2, label='Среднее')
histplot.axvline(lower_bound, color='r', linestyle='--', lw=2, label='Нижняя граница (3 σ)')
histplot.axvline(upper_bound, color='g', linestyle='--', lw=2, label='Верхняя граница (3 σ)')

plt.legend()
plt.show()
```
